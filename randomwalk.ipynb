```
from random import seed
from random import random
from matplotlib import pyplot
%matplotlib inline
seed(1)
# bron:  https://machinelearningmastery.com/gentle-introduction-random-walk-times-series-forecasting-python/
```

```
pyplot.figure(figsize=(10,10),dpi=100)

# init eerste waarde
log_random_walk = [100]
random_walk = [100]

for i in range(1, 5000):
    if random() < 0.5:
      log_factor = 0.99
      norm_factor = -1
    else:  
      log_factor = 1.01
      norm_factor = 1
    log_stap = log_random_walk[i-1] * log_factor
    norm_stap = random_walk[i-1] + norm_factor
    log_random_walk.append(log_stap)
    random_walk.append(norm_stap)

pyplot.plot(log_random_walk)
pyplot.plot(random_walk)

pyplot.show()
```
